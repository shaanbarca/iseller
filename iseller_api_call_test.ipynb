```python
import requests
from time import sleep
from datetime import datetime as dt

def fetch_all_orders(base_url: str, access_token: str, data: dict):
    # define headers for api call
    headers = {
        'Authorization': 'Bearer ' + access_token,
        'Content-Type': 'application/x-www-form-urlencoded'
    }
    
    # download all past data
    retries = 1
    while retries > 0:
        response = requests.post(base_url, headers=headers, data=data)
        if response.status_code == 200:
            return response
        elif response.status_code == 500:
            retries -= 1
            sleep(5)
            print(f"Retry remaining: {retries}. Retrying after 5 seconds...")
        else:
            print(f"Error fetching data: {response.status_code}")
            return response
    return "Failed after multiple retries"
```

```python
# Usage
api_url = "https://minumyukkaka.iseller.io/api/v2/GetOrders"

# define parameters
data = {
    'page': 1,
    'created_after': '2024-01-01T00:00:00',
    'includes': 'OrderDetails'
}

response = fetch_all_orders(api_url, access_token, page=1)
response.json()
```
