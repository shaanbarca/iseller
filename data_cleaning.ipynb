```python
import os
import re
import json
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
# iterate through every json file in data
root = '/Users/davidrinaldi/Documents/Projects/iseller/'
data_folder = 'data/data_copy/'
full_path = root + data_folder

counter = 0
for filename in os.listdir(full_path):
        
    # get full file path
    full_file_path = full_path + filename
    
    # create dataframe
    if counter == 0:
        with open(full_file_path, 'r') as jsonfile:
            json_data = json.load(jsonfile)
        iseller_data = pd.DataFrame(json_data['orders'])
        counter += 1
        continue
    
    # append the data
    with open(full_file_path, 'r') as jsonfile:
        json_data = json.load(jsonfile)
    json_data = pd.DataFrame(json_data['orders'])
    iseller_data = pd.concat([iseller_data, json_data], ignore_index=True)
    
    counter += 1
    
    if counter == 10000:
        break
    
# showcase data
iseller_data.info()
```

```python
# add time data
iseller_data['order_date'] = pd.to_datetime(iseller_data['order_date'])
iseller_data['hour'] = iseller_data['order_date'].dt.hour
iseller_data['day'] = iseller_data['order_date'].dt.day
iseller_data['month'] = iseller_data['order_date'].dt.month
iseller_data['week'] = iseller_data['order_date'].dt.isocalendar().week
iseller_data['Quarter'] = iseller_data['order_date'].dt.quarter
```

```python
# get location columns
iseller_data['location'] = iseller_data['outlet_name'].str.split(' - ').str[0]
```

```python
# Filter rows where 'items_list' has more than two elements
filtered_df = iseller_data[iseller_data['order_details'].apply(lambda x: len(x) > 2)]

# Show the filtered DataFrame
print(filtered_df.loc[6]['order_details'])
```

```python
# standardize naming
iseller_data['product_name'] = iseller_data['product_name'].str.lower()
```
